### GNN Model

1. Step: Linkfunction
<br>
Either use 1x1 convolution (for images) or LSTM (for videos) to compute Adjecency Matrix. Link function takes edge features as input and infers the connectivities between nodes.



### Training
##### def train(train_loader, model, mse_loss, multi_label_loss, optimizer, epoch, logger)
<br>
for each single set in train loader: convert node- & edge-features, adjacency matrix and node labels into torch.autograd.variables, which can be updated in the training process.  Aboved mentioned variables are inputs of GNN. The model can predict an adjacency matrix and node labels.
